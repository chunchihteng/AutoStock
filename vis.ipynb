# 1. Process Data

```python
stock_name, stock_list = get_stock('stock_list.txt')
digit_time = 20180306

fiance_stock = [ff.rstrip('\n') for ff in open('fiance_stock.txt', 'r').readlines()]
count = 0
recommand = []
indicator = [['MA', 10], ['MA', 20], ['MA', 60], ['EMA', 23]]
for s in stock_name:
    if s in fiance_stock: continue
    date, volume, price, date_correct = get_stock_data(stock_list, s, digit_time) 
    if date_correct is False: continue
        
    open_p, high_p, low_p, close_p = price[:,0], price[:,1], price[:,2], price[:,3]
    if volume[-1] < 2000: continue
    if close_p[-1] < 10 or close_p[-1] > 100: continue
    
    t1 = []
    for d in [200, 100, 60, 40, 20, 10]:
        r = np.array([(high_p[i] - open_p[i]) / (1.0*open_p[i]) for i in range(-min(d, len(open_p)), 0)])*100.0
        if np.mean(r) < 2.5:
            continue
        t1 += [[d, np.mean(r)]]
    if t1 != []:
        print "up", s, t1
    
    t2 = []
    for d in [200, 100, 60, 40, 20, 10]:
        r = np.array([(open_p[i] - low_p[i]) / (1.0*open_p[i]) for i in range(-min(d, len(open_p)), 0)])*100.0
        if np.mean(r) < 2.5:
            continue
        t2 += [[d, np.mean(r)]]
    if t2 != []:
        print "down", s, t2
    
    if (t1 != [] and t2 != []) or (t1 == [] and t2 == []) :
        continue
    try:
        draw_plot(date, volume/1000.0, price, 40, indicator=indicator)  
    except:
        pass
    if t1 != [] or t2 != []:
        count += 1
print count
```
